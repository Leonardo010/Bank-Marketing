```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import StandardScaler
import numpy as np
```

```python
for column in data.columns:
    if data[column].dtype == 'object':
        data = pd.get_dummies(data, columns=[column])
    else:
        pass
```

```python
pd.set_option("display.max_columns", None)
```

```python
data
```

```python
data.to_csv('data_binary_values.csv', index=False)
```

```python
x = data[list(data.drop('y', axis=1).columns)]
y = data['y']
```

```python
raw_treino_x, raw_teste_x, treino_y, teste_y = train_test_split(x, y, test_size= 0.25, stratify=y)
scaler = StandardScaler()
scaler.fit(raw_treino_x)
treino_x = scaler.transform(raw_treino_x)
teste_x = scaler.transform(raw_teste_x)
```

```python
SEED = 5
np.random.seed(SEED)
print("Treinaremos com %d elementos e testaremos com %d elementos" % (len(treino_x), len(teste_x)))

trees = RandomForestClassifier(random_state = SEED, n_estimators=500, n_jobs=-1)
trees.fit(treino_x, treino_y)
previsoes = trees.predict(teste_x)

acuracia = accuracy_score(teste_y, previsoes) * 100
print("A acurácia foi %.2f%%" % acuracia)
```
