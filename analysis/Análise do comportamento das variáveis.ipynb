```python
import scipy
```

```python
from scipy import stats
```

```python
correlation, pvalue = stats.pointbiserialr(data['Idade'], data['fez empréstimo ou não'])
```

```python
print(f'correlação: {correlation:.4f}')
print(f'valor P: {pvalue:.4f}')
```

Correlação entre variáveis quantitativas e variável alvo.

```python
num_columns = data.select_dtypes(include=['int64', 'float64']).columns

correlations_num = []

for col in num_columns:
    corr, pvalue = stats.pointbiserialr(data[col], data['fez empréstimo ou não'])
    correlations_num.append({'feature': col, 'correlation': corr, 'pvalue': pvalue})

correlations_num_df = pd.DataFrame(correlations_num)

correlations_num_df
```

```python
table = pd.crosstab(data['Trabalho'], data['fez empréstimo ou não'])
```

```python
chi2, p, dof, expected = stats.chi2_contingency(table)
```

```python
print(f"Qui-quadrado: {chi2}\np-value: {p}")
```

```python
obj_columns = data.select_dtypes(include='object').columns
correlation_obj = []

for col in obj_columns:
    table = pd.crosstab(data[col], data['fez empréstimo ou não'])
    chi2, p, dof, expected = stats.chi2_contingency(table)
    correlation_obj.append({'feature': col, 'chi2': chi2, 'p': p, 'dof': dof, 'expected': expected})
    
correlations_obj_df = pd.DataFrame(correlation_obj)
pd.set_option('display.max_colwidth', None)
correlations_obj_df
```
